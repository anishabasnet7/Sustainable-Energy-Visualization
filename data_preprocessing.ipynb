```python
import pandas as pd
```

```python
df = pd.read_csv("data/source_data.csv")
df.head()
```

```python
df.columns
```

```python
df.rename({
    'Entity': 'country', 
    'Year': 'year',
    'Access to electricity (% of population)':'access_to_electricity',
    'Access to clean fuels for cooking': 'access_to_clean_fuels_for_cooking',
    'Renewable-electricity-generating-capacity-per-capita': 'installed_renewable_energy_capacity_per_person',
    'Financial flows to developing countries (US $)': 'financial_flows_to_developing_countries',
    'Renewable energy share in the total final energy consumption (%)': 'renewable_energy_consumption_percentage',
    'Electricity from fossil fuels (TWh)': 'electricity_from_fuels',
    'Electricity from nuclear (TWh)': 'electricity_from_nuclear',
    'Electricity from renewables (TWh)': 'electricity_from_renewables',
    'Low-carbon electricity (% electricity)': 'low_carbon_electricity_percentage',
    'Primary energy consumption per capita (kWh/person)': 'energy_consumption_per_person',
    'Energy intensity level of primary energy (MJ/$2017 PPP GDP)': 'energy_intensity',
    'Value_co2_emissions_kt_by_country': 'co2_emissions_per_person',
    'Renewables (% equivalent primary energy)': 'renewables_percentage',
    'gdp_growth': 'gdp_growth',
    'gdp_per_capita': 'gdp_per_capita',
    'Density\\n(P/Km2)': 'population_density',
    'Land Area(Km2)': 'land_area',
    'Latitude': 'latitude',
    'Longitude': 'longitude',
}, axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Remove columns with high number missing values

df.drop(columns=[
    'financial_flows_to_developing_countries',
    'renewables_percentage',
    'installed_renewable_energy_capacity_per_person'
], inplace=True)
```

```python
# Since the missing values in the access_to_electricity column likely correspond to a lack of data from the early 2000s, we will simply replace the null values with zeros.

df["access_to_electricity"] = df["access_to_electricity"].fillna(0)
```

```python
# Since the missing values occur in the early years for the country French Guiana, 
# and the columns represent attributes that change little or not at all (e.g., latitude and longitude), 
# we can use forward fill (ffill) to propagate the last valid observation to the subsequent missing entries.

df["population_density"] = df["population_density"].ffill()
df["land_area"] = df["land_area"].ffill()
df["latitude"] = df["latitude"].ffill()
df["longitude"] = df["longitude"].ffill()
```

```python
# Since the missing values are spread across various countries and years, we can apply backfilling to populate the missing entries.

df["access_to_clean_fuels_for_cooking"] = df["access_to_clean_fuels_for_cooking"].bfill()
df["renewable_energy_consumption_percentage"] = df["renewable_energy_consumption_percentage"].bfill()
df["energy_intensity"] = df["energy_intensity"].bfill()
df["co2_emissions_per_person"] = df["co2_emissions_per_person"].bfill()
df["gdp_growth"] = df["gdp_growth"].bfill()
df["gdp_per_capita"] = df["gdp_per_capita"].bfill()
```

```python
df[df["low_carbon_electricity_percentage"].isnull()]
```

```python
# Since the null values in the electricity_from_renewables column correspond to only Tuvalu and Bermuda, 
# we can interpret this as the country not generating electricity from renewables 
# but relying on other energy sources. 
# or there is no data from the country
# So, we leave it as NULL

df[df["low_carbon_electricity_percentage"].isnull()]
```

```python
# Since the null values in the electricity_from_renewables column correspond to only Tuvalu, 
# we can interpret this as the country not generating electricity from renewables 
# but relying on other energy sources. 
# or there is no data from the country
# So, we leave it as NULL

df[df["electricity_from_renewables"].isnull()]
```

```python
# Since the null values in the electricity_from_nuclear column correspond to only Chile, Indonesia, and Tuvalu, 
# we can interpret this as the country not generating electricity from nuclear 
# but relying on other energy sources. 
# or there is no data from the country
# So, we leave it as NULL

df[df["electricity_from_nuclear"].isnull()]
```

```python
# Since the null values in the electricity_from_fuels column correspond to Tuvalu, 
# we can interpret this as the country not generating electricity from fuels 
# but relying on other energy sources. 
# or there is no data from the country
# So, we leave it as NULL

df[df["electricity_from_fuels"].isnull()]
```
